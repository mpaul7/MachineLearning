##### Load data - Pima Indian Diabetes Dataset
 This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. 
The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset.

**Description of the dataset**
- **'preg'** - Number of times pregnant
- **'plas'** - Plasma glucose concentration 
- **'pres'** - Diastolic blood pressure (mm Hg)
- **'skin'** - Triceps skin fold thickness (mm)
- **'test'** - 2-hour serum insulin
- **'mass'**- body mass index
- **'pedi'** - diabetes pedigree function
- **'age'** - age (years)
- **'class'** - outcome (diabetes - yes - (1), No - (0))

## Descriptive Statistics
Descriptive statistics can give you great insight into the shape of each attribute. 
The **describe()** function lists 8 statistical properties of each attribute. 
- Count
- Mean
- Standard Deviation
- Minimum Value
- 25th percentile

```python
# Check Statistical properties of each attribute
print(data.describe())
```

## Class Distribution
- In classification problem you need to know how balanced the class values are.
- **Imbalanced problems** - a lot more observations for one class than another
- **Balanced problems** - all class distributions are nearly the same or somewhat comparable. 
- For example: Imbalanced problems - Credit card fraud detection 

```python
# for classification problem check class distribution
class_counts = data.groupby('class').size()
print(class_counts)
```

##### Class distribution description
The output above shows that the dataset has **500** rows for **class '0'**, and **268** rows for **class '1'**
